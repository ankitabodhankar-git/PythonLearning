```python
# creating list - list can store multiple values
my_list = [10,"python", 3.6]
print(my_list)
```

```python
# indexing of list
marks= [ 10,50,66,89,45]
print([0])    #10
print([1])    #50
print([2])    #66
print([3])    #89
print([4])    #45
```

```python
# negative indexing
m = [23,56]
print([-1])  #56 last ele
print([-2])
```

```python
# list slicing 
num = [1,3,6,7,4]
print(num[1:4])  #index 1-4 mdhil
print(num[:3])   #index from 0-2
print(num[2:])   #index 2 pasun pudh
```

```python
# list with loop
# direct loop
num = [10,20,30,40,50]
for x in num:
    print(x)
```

```python
# loop using index list
num = [10,20,30,40,50]
for x in range(len(num)):
    print(num[x])
```
