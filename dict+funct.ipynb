```python
#  Write a function that takes a dictionary of products & prices and returns total bill
def total_bill(products):
    total = 0
    for price in products.values():
        total += price
    return total
items = {"pen":15,"pencils":44,"sheets":50,"bag":67}
print(total_bill(items))    
```

```python
# Write a function that takes a list of marks and returns average marks.
# avg= (45,89,66,33,12)/4 formula think then code
def average_marks(marks):
    total = 0
    for m in marks:
        total += m # m means one mark at a time
    average = total / len(marks)
    return average
marks_list = [70, 80, 60, 90]
print(average_marks(marks_list))
```

```python
# Write a function that takes employee salarys dictionary and returns employee with max salary.
def emp_salary(employees):
    return max(employees.values()) #.values use bcoz to get int values salary,marks like this
salary = {"ankita":45500, "asmita":34000, "archna":54560}
print(emp_salary(salary))

# method for name and salry
def emp_with_max_salary(employees):
    max_salary = 0
    max_employee = ""

    for name, salary in employees.items():
        if salary > max_salary:
            max_salary = salary
            max_employee = name
    return max_employee, max_salary
salary = {"ankita": 45500, "asmita": 34000, "archna": 54560}
print(emp_with_max_salary(salary))
```
