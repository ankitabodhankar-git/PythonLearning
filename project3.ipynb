```python
# student_marks_calculator.py
def calculate_total(marks):
    total = 0
    for m in marks:
        total = total + m
    return total
def calculate_average(total, count):
    return total / count
    
def check_result(avg):
    if avg >= 40:
        return "Pass"
    else:
        return "Fail"

name = input("Enter student name: ")

marks = []
for i in range(3):
    m = int(input(f"Enter marks of subject {i+1}: "))
    marks.append(m)

total = calculate_total(marks)
average = calculate_average(total, len(marks))
result = check_result(average)

print("\n Student Result : ")
print("Name:", name)
print("Total Marks:", total)
print("Average:", average)
print("Result:", result)
```
