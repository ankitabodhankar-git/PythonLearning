```python
# FUNCTION WITH MULTIPLE RETURN VALUES 
def result(marks):
    if marks >= 40:
        return "pass",marks
    else:
        return "fail",marks
status, m = result(23)
print(status, m)
```

```python
# DEFAULT ARGUMENT FUNCTION means arguments already has a values
def greet(name="student"):
    print("hello",name)
greet()
greet("ankita") #already store in db
```

```python
# KEYWORD ARGUMENT FUNCTION passes using name
def student(name,age):
    print(name,age)
student(age=22, name= "ankita")
```

```python
# VARIABLE LENGTH
# *args-multiple values
def total(*nums):
    return sum(nums) #nums become tuple
print(total(10,20,30))
```

```python
# **kwargs-key value
def info(**data):
    print(data) #data becomes dictionary
info(name="ankita", age= 20)
```

```python
# RECURSIVE FUNCTION -function that call itself
def factorial(n):
    if n==1:
        return 1
    return n*factorial(n-1)
print(factorial(8))
```

```python
# LAMBDA FUNCTION - also called anonymous means one line funct without name
# syntax- lambda arguments:expression
square = lambda x:x*x
print(square(9))  #use for filter,sorting data
```

```python
# NESTED FUNCTION - fuct inside another funct
def outer():
    def inner():
        print("Inside inner")
    inner()
outer()
```
