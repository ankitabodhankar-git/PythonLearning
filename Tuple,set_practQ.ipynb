```python
# Q  TUPLES = Store days of the week in a tuple and print only working days.
days = ("Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday")
# Print only working days (Monday to Friday)
for day in days:
    if day != "Saturday" and day != "Sunday":
        print(day)
```

```python
# SETS Practices Q
# A classroom has duplicate roll numbers by mistake. Use set to clean the data
rolls = [1, 2, 2, 3, 4, 4]
clean_rolls = set(rolls)
print(clean_rolls)
```

```python
# Given two sets of skills of two candidates, find: common skills skills only candidate A has
A = {"Python", "SQL", "HTML"}
B = {"Python", "Java"}
print("Common skills:", A & B)
print("Only A skills:", A - B)
```

```python
# Convert a list of phone numbers into a set and count unique numbers.
phones = [999, 888, 999, 777]
unique_phones = set(phones)
print("Unique count:", len(unique_phones))
```

```python
# Find users who logged in today but not yesterday using set difference.
today = {"A", "B", "C"}
yesterday = {"B", "D"}
print("New users:", today - yesterday) #practise
```

```python
# mix ques
# Store fixed admin roles in a tuple and user-assigned roles in a set. Check if user has any admin role.
admin_roles = ("admin", "manager")
user_roles = {"editor", "admin"}
for role in user_roles:
    if role in admin_roles:
        print("User has admin access")
```

```python
 # Use a tuple for fixed categories and a set for selected categories. Find invalid selectio ("S", "M", "L")
categories = ("Electronics", "Clothing", "Books")
selected = {"Electronics", "Toys"}
invalid = selected - set(categories)
print("Invalid:", invalid)
```

```python
# Store fixed product sizes in a tuple and customer-selected sizes in a set. Validate order
sizes = ("XS","S", "M", "L","XL")
order = {"M", "XXL"}
invalid = order - set(sizes)
print("Invalid size:", invalid)
```

```python
# Use a tuple for allowed file extensions and check uploaded files using set logic
allowed = ("jpg", "png", "pdf")
files = {"jpg", "exe"}
invalid = files - set(allowed)    #practise
print("Invalid files:", invalid)
```
