```python
# Write a function that takes a list and returns the count of elements
def total_count(lst):
    return len(lst) #len means count 
print(total_count([56,78,45,33,11]))

# method2
def count_elements(lst):
    count = 0
    for _ in lst: # _ this bcoz its only count or repition works
        count +=1
        return count
print(count_elements([56,78,45,33,11]))
```

```python
#  Write a function that takes a list of marks and returns average marks
def avg_marks(marks):
    total = 0
    for m in marks:
        total+=m
    return total/len(marks)
print(avg_marks([56,78,45,33,11])) #hard
```

```python
#  Write a function that takes a list and returns a new list of even numbers
def even_numbers(lst):
    even_list = [] #empty list
    for num in lst:
        if num %2==0: 
            even_list.append(num) #add list 
    return even_list #print only even num
print(even_numbers([56,78,45,33,11,4,7,88])) #hard 
```
