```python
# shopping bill generator
def total_bill(prices):
    total = 0
    for p in prices:
        total += p
    return total
prices = []
for i in range(3):
    price = int(input(f"Enter price of item {i+1}: "))
    prices.append(price)

bill = total_bill(prices)
print("Total Bill:", bill)
```
