```python
# list methods
# append() Add ONE new element at the end of the list.
list = [10,20,30,40]
list.append(50)
print(list)
```

```python
tasks = []
tasks.append("Study Python")
print(tasks)
```

```python
# extend() Add multiple elements from another list.used to merge list
a= [1,2]
a.extend([3,4])
print(a)
```

```python
# insert()  add at specific position using index
nums= [10,20,30]
nums.insert(3,40)
print(nums)
```

```python
# remove() remove by value in already in list
nums= [10,20,30]
nums.remove(20)
print(nums)
```

```python
# pop() used when deleting last or specific index
nums= [10,20,30]
nums.pop(2)
print(nums)
```

```python
# clear() delete everything from list
nums= [10,20,30, 40,50]
nums.clear()
print(nums)
```

```python
# index() — find position
nums= [10,20,30, 40,50]
print(nums.index(30))
```

```python
# count() — count repetitions
nums = [1, 2, 2, 3]
print(nums.count(2))
```

```python
# sort() arrange list
nums= [1,6,9,4,6,2]
nums.sort()
print(nums)
```

```python
# reverse()  reverse list/order
nums= [1,6,9,4,6,2]
nums.reverse()
print(nums)
```

```python
# copy()
nums= [1,6,9,4,6,2]
new = nums.copy()
print(new)
```
