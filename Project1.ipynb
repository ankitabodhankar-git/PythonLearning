```python
# PROJECT 1 
# STUDENT_RESULT_SYSTEM #think step by step
def get_result(marks):
    if marks >=35:
        return "pass"
    else:
        return "fail"

def get_grade(marks):
    if marks >= 80:
        return "A"
    elif marks >= 60:
        return "B"
    elif marks >= 40:
        return "C"
    else:
        return "Fail"
        
name = input("enter ur name :")
marks = int(input("enter ur marks:"))

result = get_result(marks)
grade = get_grade(marks)
print("\n student result")
print("name:", name)
print("marks:", marks)
print("grades:", grade)
print("result:",result)    
```
