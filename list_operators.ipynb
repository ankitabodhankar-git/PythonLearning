```python
# list operator
# concatenation means combine 2 string/list into one use + opeartor
a= [1,6,4]
b= ["ankita", "anuja", "tanuja"]
print(a + b)
```

```python
# repetition use*
x = [1,4,7]
print(x*5)
```

```python
# length use len() method
x = [ 'A', 'B', 'C','D']
print(len(x))
```

```python
# membership in , not in
x = [1,4,7,8,4,6]
if 7 in x:
    print("found")
else:
    print("not found")
```

```python
# not in 
x = [1,4,7,8,4,6]
if 7 not in  x:
    print("found")
else:
    print("not found")
```
