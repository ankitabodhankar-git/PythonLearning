```python
# Dictionary - A dictionary stores data in key : value form so we can access values by name, not position.
# always use {} curly braces
# use in real life like profile data store,setting pg,ecommer price/product,
# When do you use dictionary in Python? // dictionary when data has key–value meaning, like user profiles, login systems, product prices, API responses, or configuration settings.”
data = {"name": "ankita", "age":"22" , "marks":"95"}
print(data["name"])
```

```python
# operations
# add new data 
data = {"name": "ankita", "age":"22" , "marks":"95"}
data["hobby"] = "reading"
print(data)
```

```python
# update data
data = {"name": "ankita", "age":"22" , "marks":"95"}
data["marks"] = "89"
print(data)
```

```python
# delete data
data = {"name": "ankita", "age":"22" , "marks":"95"}
del data["age"] 
print(data)
```

```python
# get() function  
# dict[key] or safer dict.get(key) impppp
data = {"name": "ankita", "age":"22" , "marks":"95"}
print(data.get("age", "Not Found"))
print(data.get("hobby", "Not Found"))
```

```python
# loop dictionary
data = {"name": "ankita", "age":"22" , "marks":"95"}
for k in data:  # k means key
    print(k)
for v in data.values():  #v values remember this //Use .values() for sum, max, count,total,bill
    print(v)
```

```python
# loop key + value //Use .items() for both key & value
data ={'name': 'ankita', 'age': '22', 'marks': '95', 'hobby': 'reading'}
for k , v in data.items():
    print(k,v)
```

```python
# practice question 
# total marks
marks = {"math":88, "java":90, "python":78, "php":45}  #when making bill/total dont use"" on int/ numbers otherwise it will give error
total = 0
for m in marks.values():
   total += m
print(total)
```

```python
# highest marks find out
marks = {"math":88, "java":90, "python":78, "php":45}
highest = 0
for m in marks.values():
    if m > highest:
        highest = m
print(highest)
```

```python
# Create a phone book dictionary and search a number by name
phone_book = {"Ankita": "9876543210","Rahul": "9123456780","Sneha": "9988776655"}
name = input("Enter name to search: ")
number = phone_book.get(name)
if number:
    print("Phone number:", number)
else:
    print("not found")
```

```python
# Create employee salary dictionary and find employee with maximum salary
emp = {"ankita":45000,"aditya":75600,"pankaj":35600}
max_salary = 0
max_emp = ""
for name, salary in emp.items():
    if salary > max_salary:
        max_salary = salary
        max_emp = name

print("Emp with highest salary:", max_emp)
print("Salary:", max_salary)
```
